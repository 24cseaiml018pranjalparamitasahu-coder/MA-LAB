```python
import numpy as np
arr=np.array([1,2,3,4,5])
print("The array is:",arr)
```

```python
arr=np.array([1,2,3,4,5])
addition=arr+2
multiplication=arr*2
exponentiation=arr**2
print("The array is:",arr)
print("The addition is:",addition)
print("The multiplication is:",multiplication)
print("The exponentiation is:",exponentiation)
```

```python
arr=np.array([1,2,3,4,5])
ind1=arr[3]
sli1=arr[3:5]
sli2=arr[0:-2]
print("The array is:",arr)
print("value:",ind1)
print("value:",sli1)
print("value:",sli2) 
```

```python
arr1=np.array([[1,2],[4,5]])
arr2=np.array([[7,8],[10,11]])
mul=np.dot(arr1,arr2)
print("1st matrix is:",arr1)
print("2nd matrix is:",arr2)
print("Multiplication of both matrix matrix is:",mul)
```

```python
arr1=np.array([1,2])
arr2=np.array([7,8])
concatenate=np.concatenate((arr1,arr2),axis=0)
h_stack=np.hstack((arr1,arr2))
v_stack=np.vstack((arr1,arr2))
print("1st matrix is:",arr1)
print("2nd matrix is:",arr2)
print("Concatenation of both matrix matrix is:",concatenate)
print("Horizontally stacked two arrays:",h_stack)
print("Vertically stacked two arrays:",v_stack)
```

```python
arr1=np.array([1,2])
arr2=np.array([7,8])
equal= arr1==arr2
greater= arr1>arr2
less= arr1<arr2
print("1st matrix is:",arr1)
print("2nd matrix is:",arr2)
print("Comparing equal:",equal)
print("Comparing greater:",greater)
print("Comparing less:",less)
```
